## **SIZE Column**

1. Convert KB into M
2. Remove M from all values
3. Handle `Varies with Devise`

```python
df['Size'].isnull().sum()
```

```python
df['Size'].value_counts()
```

```python
# Converting KB to M and then Removing M also to make the value Nummeric

def convert_size(size):

    if 'M' in size:
        return float(size.replace('M', ''))
    elif 'KB' in size:
        return round(float(size.replace('KB', '')) / 1024, 2)
    else:
        return None 
    
df['Size'] = df['Size'].apply(convert_size)

print(df.head())
```

```python
df.sample(20)
```

## **INSTALL COLUMN**

```python
df['Installs'].value_counts()
```

```python
# Function to remove commas and plus signs, then convert to integer
def remove_commasplus(ch):
    if isinstance(ch, str):
        # Remove both commas and plus signs
        ch = ch.replace(',', '').replace('+', '')
        return int(ch)
    else:
        # Handle non-string values (like NaN or float) by returning None or 0
        return 0  # or return None if you prefer

# Apply the function to the Installs column
df['Installs'] = df['Installs'].apply(remove_commasplus)

# Display the first few rows of the modified dataframe
df.sample(20)
```
