```python
#import all libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import titanic dataset from seaborn
titanic = sns.load_dataset('titanic')
```

```python
#create a plot of the age distribution of the age and fear of the passengers
sns.histplot(titanic['age'], kde = False, bins = 30)
```

```python
titanic.plot(kind='scatter', x='age', y='fare', c='pclass', cmap='plasma')

'''

c='pclass':
The color of each data point corresponds to the passenger class (pclass). This is a categorical variable (likely with values 1, 2, 3), 
and the c parameter is used to represent different classes in different colors.

cmap='plasma':
cmap stands for colormap. It controls the color scheme of the plot. Here, the 'plasma' colormap is applied, 
meaning the data points will be colored according to the passenger class, with the colors from the 'plasma' color scale.

'''
```

```python
titanic.head()
```

```python
titanic.groupby('sex')['age'].mean().plot(kind='bar')
plt.title('Average Age by Sex in Titanic')
plt.ylabel('Average Age')
plt.xlabel('Gender')
```

```python
#bar plot using seaborn library
sns.barplot(data = titanic, x =  'survived', y = 'fare')
```

```python
sns.barplot(data = titanic, x =  'survived', y = 'fare', hue='who')
```

```python
sns.scatterplot(data = titanic, x =  'age', y = 'fare', hue='class')
plt.title('Fare vs Age by Class')
```

```python
#make scatter plot via plotly
import plotly.express as px
fig = px.scatter(titanic, x='age', y='fare', color='sex')

# Add the title
fig.update_layout(title="Age vs. Fare of Titanic Passengers, Colored by Gender")

fig.show()
```
