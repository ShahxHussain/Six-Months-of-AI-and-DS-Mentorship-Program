```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
# make two colums one of 3 random names and height
names = ['ALi', 'Ahmad', 'Hamza']
height = [5.5, 6.0, 5.8]

import plotly.express as px

#create a plot (line chart)
fig = px.line(x=names, y=height, title='Height of 3 people')
fig.show()
```

```python
px.bar(x=names, y=height, title='Height of 3 people')

#save the figure in png in output folder
fig.write_image('output/height.png')

#save as standalone html file
fig.write_html('output/height.html')
```
