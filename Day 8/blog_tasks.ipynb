# 1. Indexing & Selecting Data

```python
# .loc[] for label-based indexing
print("Selecting specific rows and columns using .loc[]")
print(data.loc[0:5, ['AREA (sq.km)', 'ALL SEXES (RURAL)']])
```

```python
# .iloc[] for position-based indexing
print("\nSelecting specific rows and columns using .iloc[]")
print(data.iloc[0:5, 0:2])
```

# Conditional selection

```python
print("\nConditional selection: Rows where 'ALL SEXES (RURAL)' > 100000")
print(data[data['ALL SEXES (RURAL)'] > 100000])
```

# 2. Data Cleaning 🧹

```python
# Handling Missing Data
print("\nHandling Missing Data")
# Drop rows with any missing values
data_cleaned = data.dropna()
print("Rows with missing data dropped:", data_cleaned.shape)

# Fill missing values with a specific value (e.g., 0)
data_filled = data.fillna(0)
print("Missing data filled with 0:", data_filled.shape)
```

# Data Type Conversion

```python
print("\nData Type Conversion")
# Convert a column to a specific data type
data['AREA (sq.km)'] = data['AREA (sq.km)'].astype(float)
print("Converted 'AREA (sq.km)' to float")

# Removing Duplicates
print("\nRemoving Duplicates")
data_no_duplicates = data.drop_duplicates()
print("Removed duplicates:", data_no_duplicates.shape)
```

# 3. Data Manipulation & Analysis 📈

```python
# Aggregation
print("\nAggregation")
# Group by a column and aggregate
grouped = data.groupby('AREA (sq.km)').agg({'ALL SEXES (RURAL)': 'sum'})
print("Aggregated data by 'AREA (sq.km)':")
print(grouped.head())

# String Operations
print("\nString Operations")
# Example of string operations on a Series
data['DISTRICT'] = data['DISTRICT'].str.lower()
print("Converted 'District' names to uppercase")

# Merging, Joining, and Concatenating
print("\nMerging, Joining, and Concatenating")
# Example of concatenating two DataFrames
df_concat = pd.concat([data, data], axis=0)
print("Concatenated DataFrame with itself")

# Reshaping Data
print("\nReshaping Data")
# Pivot example
pivoted = data.pivot_table(index='AREA (sq.km)', columns='DISTRICT', values='ALL SEXES (RURAL)', aggfunc='sum')
print("Pivoted data:")
print(pivoted.head())
```
